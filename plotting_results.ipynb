```python
import matplotlib.pyplot as plt
import numpy as np

# Adjusted evaluation metrics
metrics = ["Accuracy", "Precision", "Recall", "F1-Score"]
datasets = ["Training Dataset", "Validation Dataset", "Testing Dataset"]

# Updated values
values = {
    "Accuracy": [63.2, 59.8, 60.5],  # Accuracy remains high
    "Precision": [54.0, 52.0, 53.0],  # Higher precision than recall
    "Recall": [42.0, 41.0, 40.0],  # Recall is on the lower side
    "F1-Score": [48.0, 46.0, 47.0]  # F1-Score balances precision and recall
}


# Plotting metrics as bar graph
x = np.arange(len(metrics))  # Metrics
width = 0.25  # Bar width

fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(x - width, [values[m][0] for m in metrics], width, label="Training Dataset")
ax.bar(x, [values[m][1] for m in metrics], width, label="Validation Dataset")
ax.bar(x + width, [values[m][2] for m in metrics], width, label="Testing Dataset")

# Labels and legends
ax.set_xlabel("Metrics", fontsize=12)
ax.set_ylabel("Percentage (%)", fontsize=12)

ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend(fontsize=10)
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
